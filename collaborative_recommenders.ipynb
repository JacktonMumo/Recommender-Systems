```python
import pandas
user_item_details = pandas.read_csv("file.csv")
user_item_details.head(20)
```

```python
movie_details = pandas.read_csv("Movie_Id_Titles.csv")
movie_details.head(20)
```

```python
# merging the two datasets using the merge()
data = pandas.merge(user_item_details, movie_details, on ='item_id')
data.head(20)
```

```python
# get the number of ratings per movie. groupby()
mean_rating = pandas.DataFrame(data.groupby('title') ['rating'].mean()) # creating a tabular representation of the mean ratings
mean_rating
# creating a column and storing the mean rating
mean_rating['MeanRating'] = pandas.DataFrame(data.groupby('title') ['rating'].mean())
mean_rating.head(10)

# create a column(number_of_rating) and store the count of rating
mean_rating['Count'] = pandas.DataFrame(data.groupby('title')['rating'].count())
mean_rating.head(10)
```

```python
# find which movie did each user rate, and what was the rating. we will use pivot_table()
# pivot_table() - it is an interactive way to quickly summarize large amounts of data
my_pivot = data.pivot_table(values='rating', index='user_id', columns='title')
my_pivot.head(5)
```

```python
# we will find the relationship between different movies, using correlation, so that if a user has highly rated the movie we can recommend movies with similar correlation to them
selected_movie = my_pivot['12 Angry Men (1957)']
# find similar movies to this movie based on ratings. corr(qualitative). corrwith(quantitative)
similar = my_pivot.corrwith(selected_movie)
similar_dataframe =pandas.DataFrame(similar,columns=['correlation'])# creating a DF that has a new column(correlation) containing the correlation btwn the selected movie and other movies
similar_dataframe.sort_values('correlation',ascending=False).head(10)
```

```python
import pandas 
data = pandas.read_csv('shared_articles.csv')
data
```

```python
data.isnull().sum()
```

```python
data.drop(['Unnamed: 13'], axis=1,inplace=True)
data.drop(['Unnamed: 14'], axis=1,inplace=True)
data.drop(['Unnamed: 15'], axis=1,inplace=True)
data.isnull().sum()
```

```python
# fill 'unknown' in null cell in column authorUserAgent, authorRegion, and authorCountry
data['authorUserAgent'].fillna('unknown', inplace=True)
data['authorRegion'].fillna('unknown', inplace=True)
data['authorCountry'].fillna('unknown', inplace=True)
data.isnull().sum()
```

```python
# replace en with english in column lang
data['lang'].replace({
    'en':'English'
},inplace=True)
data.head(3)
```

```python
# we need the column text only.
X = data['text'].values
X
```

```python
# we will utilize sentence-transformers library for our NLP processing of our text, our aim is to encode our text in X into numerical feature(matrix)
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('distilbert-base-nli-stsb-mean-tokens')

embeddings = model.encode(X,show_progress_bar=True)
#  encoding is changing a message to a code eg hi => 1001/50
# show_progress_bar shows the encoding process progress
embeddings
```

```python
# pass the above array to be in tabular format
new_data = pandas.DataFrame(embeddings)
new_data
```

```python
# for us to get the similarity, we will use cosine similarity
from sklearn.metrics.pairwise import cosine_similarity # cosine similarity calculates the similarity between all pairs in a given data structure, in our case the data is in form of a matrix
cosine_data = cosine_similarity(new_data)
cosine_data
cosine_data_table = pandas.DataFrame(cosine_data)
cosine_data_table
```

```python
# now lets retrieve similar articles to a given news ID based on a pre-defined cosine similarity matrix
def recommenders(newsid):
    index = cosine_data_table.loc[newsid].sort_values(ascending=False).index.tolist()[1:10]
    # loc[] it provides a convenient way to access specific rows or columns based on their index or name
    # tolist() converts dataframe or series into a python list(not an array)
    # [1:10] means we will have the top 9 articles that are similar to the article we want(newsid)
    print("Similar Articles are:", index)

    news_titles = data['title'].loc[index].values
    # this uses similar news article id (index) to retrieve corresponding titles from the column title in the dataset data
    # the .values extracts the actual titles from the resulting dataframe
    print("Similar Articles Title are:", news_titles)

recommenders(3121)
```
